## 📝 El Reto de Clase: "Limpieza de Logs de Auditoría"

**Contexto del Problema**:

Imaginen que están construyendo la primera fase de un pipeline ETL para el área de finanzas.
El sistema de seguridad generó un registro (log) de auditoría, pero el texto viene "sucio" con
caracteres especiales en los extremos, espacios en blanco de más, y con la información
desordenada.
Su misión como ingenieros de datos es limpiar ese registro, extraer los montos y generar un
reporte limpio en pantalla.

**Instrucciones para el Alumno:**
1. Declaración de Variables: Crea dos variables obligatorias con tus datos:
○ mi_nombre: Tu primer nombre (ej. 'Angel').
○ mi_apellido: Tu primer apellido (ej. 'Tintaya').
2. El Registro Sucio: Copia exactamente la siguiente variable en tu celda de código (este
es el texto que envió el sistema):
3. Python

```python
mi_nombre = 'Juan Pablo'
mi_apellido = 'Mendoza'
```

```python
registro_sucio = " ##[ID-9875]__Monto:S/.4500__Estado:PENDIENTE##"
```

**Tareas de Limpieza y Transformación:**


*   Paso A: Quita los espacios en blanco y los caracteres ## que están en los
extremos (bordes izquierdo y derecho) del texto usando .strip().






```python
paso_a = registro_sucio.strip(" #")
print(paso_a)
```

* Paso B: Convierte todo el texto resultante a minúsculas usando .lower().

```python
paso_b = paso_a.lower()
print(paso_b)
```

* Paso C: Usa .replace() para cambiar la palabra "pendiente" por la
palabra "procesado".

```python
paso_c = paso_b.replace("pendiente","procesado")
print(paso_c)
```

* Paso D: Divide el texto limpio en partes usando como separador los guiones
bajos dobles __ mediante el método .split().

```python
paso_d = paso_c.split("__")
print(paso_d)
```

* Paso E: De la lista resultante del split, extrae la parte que corresponde al monto
e identifica en qué posición (índice) se encuentra la primera letra "m" de la
palabra "monto" usando .index().

```python
paso_e1 = paso_d[1]
print(paso_e1)
```

```python
paso_e2 = paso_e1.index("m")
paso_e2
```
