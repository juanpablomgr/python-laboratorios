**Parte 1: Ingesta y Limpieza de Cadenas (Strings)**

El sistema ha arrojado la siguiente cadena de texto correspondiente a la última transacción.
Cópiela exactamente igual en su Colab:

```python
log_crudo = "$$[ESTADO:ok]||trx_id:9934||monto:S/.1250.50||alerta:falso$$ "
```

Misiones:
1. Utilice el método adecuado para eliminar los espacios en blanco y los símbolos de dólar
$$ que se encuentran en los extremos de la cadena.
2. Convierta toda la cadena resultante a letras mayúsculas.
3. El delimitador que separa cada campo es la doble barra vertical ||. Utilice el método
correcto para dividir la cadena en partes usando este delimitador.
4. Guarde el resultado final en una variable llamada lista_campos e imprímala en pantalla
usando print().

```python
#1. Utilice el método adecuado para eliminar los espacios en blanco y los símbolos de dólar

log_stripe = log_crudo.strip("$ ")
log_stripe
```

```python
#2. Convierta toda la cadena resultante a letras mayúsculas.

log_upper = log_stripe.upper()
log_upper
```

```python
#3. El delimitador que separa cada campo es la doble barra vertical ||.
  #Utilice el método correcto para dividir la cadena en partes usando este delimitador.
log_split = log_upper.split("||")
log_split
```

```python
#4. Guarde el resultado final en una variable llamada lista_campos e imprímala en pantalla usando print().
lista_campos = log_split
print(lista_campos)
```

4. Simule un proceso de conexión de red inestable utilizando un bucle while. Configure un
contador de intentos inicializado en 1. El bucle debe imprimir "Intentando conectar al

```python
# 1. Cree un Diccionario llamado reporte_diario con las siguientes llaves y valores:

    #○ "fecha": "2026-06-07"

    #○ "lotes_procesados": Una lista con los montos [350.0, 1250.50, 90.0, 2100.0]

    #○ "operador": Ingrese su Nombre y Apellido.

reporte_diario = {"fecha" : "2026-06-07",
                  "lotes_procesados" : [350.0, 1250.50, 90.0, 2100.0] ,
                  "operador" : "Juan Pablo Mendoza"}
reporte_diario
```

```python
# 2. Utilice un bucle for combinado con la función enumerate() para recorrer la lista de lotes_procesados dentro del diccionario.

for indice, monto in enumerate(reporte_diario["lotes_procesados"], start=1) :
  print(indice, monto)
```

```python
# 3.Dentro del bucle, utilice una f-string para imprimir: "Lote N° [Índice] procesado por un valor de $[Monto]".
  #Haga que el índice comience en 1.

for indice, monto in enumerate(reporte_diario["lotes_procesados"], start=1) :
  print(f"Lote N° {indice} procesado por un valor de ${monto}")
```

```python
for indice, monto in enumerate(reporte_diario["lotes_procesados"]) :
  print(f"Lote N° {indice + 1} procesado por un valor de ${monto}")
```

```python
# 4. Simule un proceso de conexión de red inestable utilizando un bucle while.
# Configure un contador de intentos inicializado en 1. El bucle debe imprimir "Intentando conectar al servidor de destino... Intento [X]" y
# detenerse automáticamente al llegar al intento número3.

max_intentos = 3
num_intentos = 1
while num_intentos <= max_intentos:
    print(f"Intentando conectar al servidor de destino... Intento {num_intentos}")
    num_intentos += 1
```
